```python
sql = """
select * from sales
"""
_db.executeAll(sql).head(2)
```

```python
## 컬럼의 이름이나 테이블이 이름이 " " 공백이 존재하면 "`"(백쿼트) 특수문자로 묶어준다.
sql = """
select `Item Type` from sales
"""

_db.executeAll(sql)
```

```python
## Country기준으로 그룹화 Units Sold의 평균값, 합계

sql = """
select 
Country, 
AVG(`Units Sold`) as avg, 
COUNT(`Units Sold`) as cnt, 
SUM(`Units Sold`) as sum 
from sales 
group by Country
"""

_db.executeAll(sql)
```

```python
## 컬럼에서 중복 값을 제거하고 값들은 하나씩만 출력
sql = """
select 
distinct(Region) as region 
from sales
"""
_db.executeAll(sql)
```
