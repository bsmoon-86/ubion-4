### 이중 쿼리문
1. 부서의 위치가 'NEW YORK', 'DALLAS'가 아닌 소속 사원들의 정보를 출력
2. 부서의 위치가 NEW YORK 과 DALLAS 가 아닌 지역의 부서번호.
3. 사원 정보 테이블에서 위에서 찾은 부서 번호를 조건으로 사원 정보 테이블을 출력

```python
## 부서의 위치는 어느 테이블 -> dept 
## 2번 과정
sql = """
SELECT DEPTNO FROM dept WHERE LOC NOT IN ('NEW YORK', 'DALLAS')
"""
_db.executeAll(sql)
```

```python
sql = """
SELECT * FROM emp WHERE DEPTNO IN (
    SELECT DEPTNO FROM dept WHERE LOC NOT IN ('NEW YORK', 'DALLAS')
)
"""
_db.executeAll(sql)
```
