```python
## inplace 속성

df_4 = df_3.sort_values("권역") 
# df_3에 df_3데이터프레임에 '권역' 컬럼을 기준으로 오름차순 정렬을 한 데이터프레임을 할당
# 이 코드를 실행을 하면 df_3라는 데이터프레임이 변경.

df_3.sort_values("권역", inplace=True)
# df_3 = df_3.sort_values("권역")과 같은 의미
```

```python
df_3["총이윤"] = df_3["총수익"] - df_3["총비용"]
df_3.head()
```

```python
df_3.drop(["총수익", "총비용", "총이윤"], axis=1, inplace=True)
```

```python
df_3.head(2)
```

df_3 데이터프레임에서 
총수익, 총비용, 총이윤이라는 파생변수를 생성.

```python
## 총수익 -> 판매단위 * 단가
## 총비용 -> 판매단위 * 원가
## 총이윤 -> 총수익 - 총비용
df_3["총수익"] = df_3["판매단위"] * df_3["단가"]
df_3["총비용"] = df_3["판매단위"] * df_3["원가"]
df_3["총이윤"] = df_3["총수익"] - df_3["총비용"]
df_3.head(2)
```
